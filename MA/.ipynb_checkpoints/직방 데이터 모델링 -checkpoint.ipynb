```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
# 데이터를 복사하여 정규화를 진행한 데이터를 따로 만들어줌 
normalzie_data = data.copy()
```

```python
# 연속형 변수 표준화 

scaler = StandardScaler()

scaler.fit(normalzie_data[float_feature])
scaled_data = scaler.transform(normalzie_data[float_feature])

# transform()시 스케일 변환된 데이터 세트가 Numpy ndarray로 반환돼 이를 DataFrame으로 변환
normalzie_data[float_feature] = scaled_data
```

```python
# 로그 변환 
normalzie_data['transaction_real_price'] = np.log1p(data['transaction_real_price'])

data['transaction_real_price'] = np.log1p(data['transaction_real_price'])
```

```python
# 레이블 인코딩 
encoder = LabelEncoder()
encoder.fit(data['year_of_completion'])
label = encoder.transform(data['year_of_completion'])
```

```python
data['year_of_completion'] = label
normalzie_data['year_of_completion'] = label
```

```python
# 범주형 변수 이름 추출 
category_feature = [col for col in data.columns if data[col].dtypes == "object"]
```

```python
# 더미 변수 생성 
data = pd.get_dummies(data, columns=category_feature)
normalzie_data = pd.get_dummies(normalzie_data, columns=category_feature)
```

```python
X = data.drop(columns=['transaction_real_price'], axis=1)
y = data['transaction_real_price']
```

```python
# 학습데이터셋 검증데이터셋 구축 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
X_2 = normalzie_data.drop(columns=['transaction_real_price'], axis=1)
y_2 = normalzie_data['transaction_real_price']
```

```python
# 학습데이터셋 검증데이터셋 구축 
X_train2, X_test2, y_train2, y_test2 = train_test_split(X_2, y_2, test_size=0.3, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

def rmsle(y, pred):
    log_y = np.log1p(y)
    log_pred = np.log1p(pred)
    squared_error = (log_y - log_pred) ** 2
    rmsle = np.sqrt(np.mean(squared_error))
    return rmsle

def rmse(y, pred):
    return np.sqrt(mean_squared_error(y, pred))
    
    
def r_2(y, pred):
    return r2_score(y, pred)
    
    
def evaluate_regr(y, pred):
    rmsle_val = rmsle(y, pred)
    rmse_val = rmse(y, pred)
    mae_val = mean_absolute_error(y, pred)
    r_score = r_2(y, pred)
    print('R2: {0:.3f}, RMSLE: {1:.3f}, RMSE: {2:.3f}, MAE: {3:.3f}'.format(r_score, rmsle_val, rmse_val, mae_val))
```

```python
# scale을 안한 데이터 셋 테스트 데이터 성능 
lr_reg = LinearRegression()
lr_reg.fit(X_train, y_train)
pred = lr_reg.predict(X_test)

evaluate_regr(y_test, pred)
```

```python
# scale 해준 데이터 셋 테스트 데이터 성능 
scale_lr_reg = LinearRegression()
scale_lr_reg.fit(X_train2, y_train2)
scale_pred = scale_lr_reg.predict(X_test2)

evaluate_regr(y_test2, scale_pred)

# 둘다 성능 상에 차이가 없음 -> 로그변환만 수행한 기존 데이터 사용 
```

```python
# Ridge, Lasso 학습, 예측, 비교 

ridge_reg = Ridge()
ridge_reg.fit(X_train, y_train)
pred = ridge_reg.predict(X_test)
print('Ridge:')
evaluate_regr(y_test, pred)
print()
print('Lasso:')
lasso_reg = Lasso()
lasso_reg.fit(X_train, y_train)
pred = lasso_reg.predict(X_test)
evaluate_regr(y_test, pred)


# 규제 모형도 별 효과가 없다.
```

```python
from sklearn.tree import DecisionTreeRegressor

dt_reg = DecisionTreeRegressor()
dt_reg.fit(X_train, y_train)
pred = dt_reg.predict(X_test)

evaluate_regr(y_test, pred)
```

```python
# 교차 검증 

from sklearn.model_selection import cross_val_score

def get_avg_rmse_cv(models):
    
    for model in models:
        # 분할하지 않고 전체 데이터로 cross_val_score() 수행. 모델별 CV RMSE 값과 평균 RMSE 출력 
        rmse_list = np.sqrt(-cross_val_score(model, X, y, scoring='neg_mean_squared_error', cv=5))
        
        rmse_avg = np.mean(rmse_list)
        print('\n{0} CV RMSE 값 리스트: {1}'.format(model.__class__.__name__, np.round(rmse_list, 3)))
        print('{0} CV 평균 RMSE 값: {1}'.format(model.__class__.__name__, np.round(rmse_avg, 3)))
    
    
# 앞 예제에서 학습한 lr_reg, ridge_reg, lasso_reg 모델의 CV RMSE 값 출력 
models = [ridge_reg, lasso_reg]
get_avg_rmse_cv(models)
```

```python
# 모델과 학습/테스트 데이터 세트를 입력하면 성능, 평가 수치를 반환 
def get_model_predict(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    pred = model.predict(X_test)
    print('###', model.__class__.__name__, '###')
    evaluate_regr(y_test, pred)
```

```python
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
```

```python
# 랜덤 포레스트, XGBoost, LightGBM model별로 평가 수행 
rf_reg = RandomForestRegressor()

# , xgb_reg, lgbm_reg
for model in [rf_reg]:
    get_model_predict(model, X_train, X_test, y_train, y_test)
```

```python
xgb_reg = XGBRegressor()
lgbm_reg = LGBMRegressor()

for model in [xgb_reg, lgbm_reg]:
    get_model_predict(model, X_train, X_test, y_train, y_test)
```
